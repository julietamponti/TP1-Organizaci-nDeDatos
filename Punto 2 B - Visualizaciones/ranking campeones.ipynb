```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
arc = pd.read_csv('ganadores.csv')
```

```python
arc.dropna(inplace=True)
```

```python
arc
```

```python
arc.columns
```

```python
arc = arc.astype({'Jugador;cantidad ganadas': 'str'})
```

```python
arc['Jugador'] = arc['Jugador;cantidad ganadas'].apply(lambda x: x.split(';')[0])
```

```python
arc['CantidadGanadas'] = arc['Jugador;cantidad ganadas'].apply(lambda x: x.split(';')[1])
```

```python
arc.columns
```

```python
arc = arc[['Jugador', 'CantidadGanadas']]
```

```python
arc
```

```python
arc = arc.astype({'CantidadGanadas': 'int64'})
```

```python
sns.set(rc={"figure.dpi":150})
sns.set(rc = {'figure.figsize':(19.2,9.6)})
ax = sns.barplot(x="Jugador", y="CantidadGanadas", data=arc)
```
