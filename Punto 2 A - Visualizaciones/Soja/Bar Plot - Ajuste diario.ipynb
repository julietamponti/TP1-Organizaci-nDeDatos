```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dt = pd.read_csv('datasetRofex4 - datasetRofex2.csv',usecols=['Fecha','Aj.Dif.'])
dt.head(4)
```

```python
dt['Fecha'] = pd.to_datetime(dt['Fecha'])
```

```python
dt['A単o'] = dt['Fecha'].dt.year
```

```python
dt['Mes'] = dt['Fecha'].dt.month
```

```python
dt['Dia'] = dt['Fecha'].dt.day
```

```python
dt.head(2)
```

```python
dt=dt[dt['A単o']==2008]
```

```python
dt=dt[dt['Mes']==8]
```

```python
dt = dt[['Aj.Dif.','Dia']]
```

```python
sns.set(rc={"figure.dpi":200})
ax = sns.barplot(x="Dia", y="Aj.Dif.", data=dt,color='g')
ax.set_ylabel("Ajuste(U$D por tolenada)", fontsize = 8)
ax.set_xlabel("Dia (MesAgostoDel2008)", fontsize = 8)
ax.set_title("Inestabilidad del precio de la soja",size =10)
figure = ax.get_figure()    
figure.savefig('InestabilidadSoja-BarPlot.png')
```
